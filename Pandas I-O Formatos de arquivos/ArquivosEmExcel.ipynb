```python
import pandas as pd
```

```python
url = 'https://github.com/alura-cursos/Pandas/blob/main/emissoes_CO2.xlsx?raw=True'
```

```python
# Verificando se tem mais pastas no arquivo
pd.ExcelFile(url).sheet_names
```

```python
# Separando em variaveis diferentes
emissoes = pd.read_excel(url, sheet_name='emissoes_C02')
percapita = pd.read_excel(url, sheet_name='emissoes_percapita')
fontes = pd.read_excel(url, sheet_name='fontes')
```

```python
# Salvando em arquivos excel
emissoes.to_excel('emissoes_CO2.xlsx', index=False)
percapita.to_excel('percapita_CO2.xlsx', index=False)
fontes.to_excel('fontes_CO2.xlsx', index=False)
```
