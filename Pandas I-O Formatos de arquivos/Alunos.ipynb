```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/pandas-conhecendo-a-biblioteca/main/desafios/alunos.csv'
```

```python
dados = pd.read_csv(url) # Lendo o arquivo
dados.replace(7, 8, inplace=True) # Todos alunos que tiraram 7 pontos ganharam mais 1 ponto
```

```python
dados.fillna(0, inplace=True) # Tratando os valores Nan para 0
```

```python
dados.drop([7,8], axis=0, inplace=True) #Removendo Carlos e Alice
```

```python
aprovados = dados[dados['Aprovado']] # Alunos Aprovados
reprovados = dados[dados['Aprovado'] == False] # Alunos Reprovados
```

```python
# Criando os arquivos
aprovados.to_csv('alunos_aprovados.csv', index=False)
reprovados.to_csv('alunos_reprovados.csv', index=False)
```

```python
dados['Pontos_extras'] = dados['Notas'] * 0.4 # Adicionando 40% de pontos extra
dados['Notas_finais'] = dados['Notas'] + dados['Pontos_extras'] # Somando nota final
dados['Aprovado_final'] = dados['Notas_finais'].apply(lambda i: True if i >= 6 else False) # Lista de aprovados finais
dados
```

```python
dados.query('Aprovado == False & Aprovado_final == True') # Alunos que foram aprovados depois dos 40% de pontos extra
```
