```python
import sqlalchemy
import pandas as pd
```

```python
from sqlalchemy import create_engine, MetaData, Table, inspect, text
```

```python
engine = create_engine('sqlite:///:memory:') # Configurando o 'motor'
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/refs/heads/main/clientes_banco.csv'
```

```python
dados = pd.read_csv(url)
dados.head(2)
```

```python
dados.to_sql('clientes', engine, index=False) # Converte o DataFrame em uma tabela SQL chamada 'clientes'
```

```python
inspecionar = inspect(engine)
print(inspecionar.get_table_names()) # Lista todas as tabelas existentes
```

```python
# Query SQL para selecionar apenas clientes empregados
query = 'SELECT * FROM clientes WHERE Categoria_de_renda="Empregado"'
```

```python
empregados = pd.read_sql(query, engine)
```

```python
empregados.to_sql('Clientes Empregados', con=engine, index=False) # Salvando...
```

```python
pd.read_sql_table('Clientes Empregados', engine)
```

```python
query = 'DELETE FROM clientes WHERE ID_Cliente=5008804'
with engine.connect() as conexao:
    result = conexao.execute(text(query))
    conexao.commit()
```

```python
pd.read_sql_table('clientes', engine)
```
