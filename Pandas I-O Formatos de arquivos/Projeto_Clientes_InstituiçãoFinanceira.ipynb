# **1 - Criar o banco de dados local com a biblioteca SQLAlchemy.**

```python
import sqlalchemy
import pandas as pd
from sqlalchemy import create_engine, MetaData, Table, inspect, text
```

```python
# Criando motor
engine = create_engine('sqlite:///:memory:')
```

```python
url = 'https://raw.githubusercontent.com/alura-cursos/Pandas/main/clientes_banco.csv'
database = pd.read_csv(url)
database
```

# **2 - Escrever os dados do arquivo CSV neste banco de dados local.**

```python
database.to_sql('clientes', engine, index=False)
```

# **3 - Realizar três atualizações no banco de dados:**

*   Atualizar o registro do cliente de ID 6840104 que teve o rendimento anual alterado para 300000.
*   Excluir o registro do cliente de ID 5008809, pois essa pessoa não possui mais conta na instituição financeira.
*   Criar um novo registro de cliente seguindo as especificações abaixo:

    1. ID_Cliente: 6850985
    2. Idade: 33
    3. Grau_escolaridade: Doutorado
    4. Estado_civil: Solteiro
    5. Tamanho_familia: 1
    6. Categoria_de_renda: Empregado
    7. Ocupacao: TI
    8. Anos_empregado: 2
    9. Rendimento_anual: 290000
    10. Tem_carro: 0
    11. Moradia: Casa/apartamento próprio

```python
# Atualizando o cliente 6840104
with engine.connect() as conexao:
    result = conexao.execute(text('UPDATE clientes SET Rendimento_anual="300000" WHERE ID_Cliente=6840104'))
    conexao.commit()
```

```python
# Deletando o cliente 5008809
with engine.connect() as conexao:
    result = conexao.execute(text('DELETE FROM clientes WHERE ID_Cliente=5008809'))
    conexao.commit()
```

```python
# Adicionando o cliente 6850985
with engine.connect() as conexao:
    result = conexao.execute(text(f'INSERT INTO clientes (ID_Cliente, Idade, Grau_escolaridade, Estado_civil, Tamanho_familia, Categoria_de_renda, Ocupacao, Anos_empregado, Rendimento_anual, Tem_carro, Moradia) VALUES (6850985, 33, "Doutorado", "Solteiro", 1, "Empregado", "TI", 2, 290000, 0, "Casa/apartamento próprio")'))
    conexao.commit()
```
